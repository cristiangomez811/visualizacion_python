## Graficos de distribución

### Histograma 

El histograma es el grafico basico. Se usa para graficar la distribución  
numerica de una variable. recibe una variable numerica separada por interbalos  
Estos intervalos pueden ser espacíos de tiempo, se dan a la interpretacion del dataset. 

```python
# Import library and dataset
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
```

#### Histograma basico de seaborn

```python
# Make default histogram of sepal length
sns.distplot( df["sepal_length"] )
#sns.plt.show()
```

#### Histograma con control de interbalos

```python
# Control the number of bins
sns.distplot( df["sepal_length"], bins=20 )
#sns.plt.show()
```

#### Solo el histograma

```python
# Hist only
sns.distplot( a=df["sepal_length"], hist=True, kde=False, rug=False )
#sns.plt.show()
 
```

#### Histograma con rugs

```python
 
# Hist + Rug + kernel density
sns.distplot( a=df["sepal_length"], hist=True, kde=True, rug=True )
#sns.plt.show()
 
```

#### Ajustes en el rug 

```python
# To change parameters of rug
sns.distplot( a=df["sepal_length"], rug=True,
rug_kws={"color": "r", "alpha":0.3, "linewidth": 2, "height":0.2 }
)
```

#### Ajustes en la linea de distribución 

```python
# To change parameters of density distribution
sns.distplot( a=df["sepal_length"], kde=True,
kde_kws={"color": "g", "alpha":0.3, "linewidth": 5, "shade":True }
)
#sns.plt.show()
```

#### Cambiando la orientación del histograma

```python
# Vertical hist
sns.distplot( df["sepal_length"] , color="skyblue", vertical=True)
#sns.plt.show()
```

#### Histograma con boxplot 
*mas las nuevas graficas se explican mas adelante*

```python
# Cut the window in 2 parts 
# For this is necesary matplotlib
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df["sepal_length"], ax=ax_box)
sns.distplot(df["sepal_length"], ax=ax_hist)
 
# Remove x axis name for the boxplot
ax_box.set(xlabel='')
```

### BoxPlot

Un **Boxplot** es uno de los mas usados para ver la distribucion 
de una muestra de datos. El boxplot se divide en **4 secciones  
dos lineas y dos cuadros** que representan los **cuartiles** de una  
distribucion estadistica; Los cuartiles superior e inferior son  
lineas y los cuartiles centrales son cuadros separados por una  
linea que representa la **media** de la muestra. Desde el  
boxplot tambien se puede ver atipicos en los datos por medio de *

Para mas infromación sobre su interpretación consulte [Aquí](https://support.minitab.com/es-mx/minitab/18/help-and-how-to/graphs/how-to/boxplot/interpret-the-results/key-results/)

#### Agregar jitter 

El boxplot es una gran herramienta para ver la distribucion 


```python

# Usual boxplot
ax = sns.boxplot(x='species', y='sepal_length', data=df)
 
# Add jitter with the swarmplot function.
ax = sns.swarmplot(x='species', y='sepal_length', data=df, color="grey")
```

#### Agregando la cantidad de observaciones por boxplot

```python
ax = sns.boxplot(x="species", y="sepal_length", data=df)
 
# Calculate number of obs per group & median to position labels
medians = df.groupby(['species'])['sepal_length'].median().values
nobs = df['species'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]
 
    
# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
    horizontalalignment='center', size='x-small', color='w', weight='semibold')

#sns.plt.show()
```

## Graficos de Correlación

### Scatter plot

Un scatterplot es una representación en dos dimenciones de dos variables  
númericas o categoricas. Un scatterplot crea en el espacio bi dimencional  
un punto por cada observación en el dataset. Este grafico se usa mucho para  
hallar visualmente correclaciones o agrupaciones entre variables en un dataset

#### Scatterplot basico

```python
# Without regression fit:
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], fit_reg=False)
#sns.plt.show()
```

#### Scatterplot con regression

```python
# use the function regplot to make a scatterplot
sns.regplot(x=df["sepal_length"], y=df["sepal_width"])
#sns.plt.show()
```

#### Configuración de los marcadores

```python
# Change shape of marker
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], marker="+", fit_reg=False)
#sns.plt.show()
```

#### Configurando color, tamaño y transaparencia

```python
# More marker customization:
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], fit_reg=False, scatter_kws={"color":"darkred","alpha":0.3,"s":200} )
#sns.plt.show()
```

#### Configurando la linea de regresion

```python
# plot
sns.regplot(x=df["sepal_length"], y=df["sepal_width"], line_kws={"color":"r","alpha":0.7,"lw":5})
#sns.plt.show()
```

#### Configurando colores por categorias

```python
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
 
#sns.plt.show()
```

#### Cambiando el marcador por categoria

```python

# give a list to the marker argument
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, markers=["o", "x", "1"])
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
 
#sns.plt.show()
```

#### Cambiando la paleta de colores

```python
# Use the 'palette' argument
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, palette="Set2")
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
 
```

#### Ajustando manualmente los colores de las categorias

```python
# Provide a dictionary to the palette argument
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, palette=dict(setosa="#9b59b6", virginica="#3498db", versicolor="#95a5a6"))
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
```

#### Control en los limites del grafico

```python
# basic scatterplot
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False)
 
# control x and y limits
plt.ylim(0, 20)
plt.xlim(0, None)
```

#### Asignando un color a cada marcador

```python
# Create data frame. The last column is the needed color
df = pd.DataFrame(np.random.random((100,2)), columns=["x","y"])
 
# Add a column: the color depends of x and y values, but you can use whatever function.
value=(df['x']>0.2) & (df['y']>0.4)
df['color']= np.where( value==True , "#9b59b6", "#3498db")
 
# plot
sns.regplot(data=df, x="x", y="y", fit_reg=False, scatter_kws={'facecolors':df['color']})
#sns.plt.show()
```

#### Haciendo anotaciones en el scatter

```python
# Create dataframe
df = pd.DataFrame({
'x': [1, 1.5, 3, 4, 5],
'y': [5, 15, 5, 10, 2],
'group': ['A','other group','B','C','D']
})
 
# basic plot
p1=sns.regplot(data=df, x="x", y="y", fit_reg=False, marker="o", color="skyblue", scatter_kws={'s':400}) 

# add text annotation
p1.text(3+0.2, 4.5, "An annotation", horizontalalignment='left', size='medium', color='black', weight='semibold')
#sns.plt.show()

```

#### Haciendo anotaciones en cada marcador

```python
# basic plot
p1=sns.regplot(data=df, x="x", y="y", fit_reg=False, marker="o", color="skyblue", scatter_kws={'s':400})
 
# add annotations one by one with a loop
for line in range(0,df.shape[0]):
     p1.text(df.x[line]+0.2, df.y[line], df.group[line], horizontalalignment='left', size='medium', color='black', weight='semibold')

#sns.plt.show()
```

### Scatter plot conectado

Este es un tipo diferente scatter plot, funciona de una   
forma similar pero a diferencia del basico este tiene  
cada punto de la frafica conectado al siguiente

#### Scatter plot conectado basico

```python
# data
df=pd.DataFrame({'x': range(1,10), 'y': np.random.randn(9)*80+range(1,10) })
 
# plot
plt.plot( 'x', 'y', data=df, linestyle='-', marker='o')
plt.show()
```

#### Estilos de linea

```python
plt.plot( 'x', 'y', data=df, linestyle='dashed')
plt.show()

plt.plot( [1,1.1,1,1.1,1], linestyle='-' , linewidth=4)
plt.text(1.5, 1.3, "linestyle = '-' ", horizontalalignment='left', size='medium', color='C0', weight='semibold')
plt.plot( [2,2.1,2,2.1,2], linestyle='--' , linewidth=4 )
plt.text(1.5, 2.3, "linestyle = '--' ", horizontalalignment='left', size='medium', color='C1', weight='semibold')
plt.plot( [3,3.1,3,3.1,3], linestyle='-.' , linewidth=4 )
plt.text(1.5, 3.3, "linestyle = '-.' ", horizontalalignment='left', size='medium', color='C2', weight='semibold')
plt.plot( [4,4.1,4,4.1,4], linestyle=':' , linewidth=4 )
plt.text(1.5, 4.3, "linestyle = ':' ", horizontalalignment='left', size='medium', color='C3', weight='semibold')
plt.axis('off')
plt.show()
```

#### Ajustando el grosor de la linea

```python
plt.plot( 'x', 'y', data=df, linewidth=22)
plt.show()
```

#### Conbinando diferentes tipos de linea en un plot

```python
# Data
df=pd.DataFrame({'x': range(1,11), 'y1': np.random.randn(10), 'y2': np.random.randn(10)+range(1,11), 'y3': np.random.randn(10)+range(11,21) })
 
# multiple line plot
plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'x', 'y2', data=df, marker='', color='olive', linewidth=2)
plt.plot( 'x', 'y3', data=df, marker='', color='olive', linewidth=2, linestyle='dashed', label="toto")
plt.legend()
```
